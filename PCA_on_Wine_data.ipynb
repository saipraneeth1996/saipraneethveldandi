# Principal Component Analysis:

```python
from sklearn.decomposition import PCA
```

For n_components=13

```python
pca13 = PCA(n_components=13)
pc13 = pca13.fit_transform(X_std)
pc13_df = pd.DataFrame(data = pc13)
```

```python
pc13_df.head()
```

Explained variance for each component

```python
ev=pca13.explained_variance_
ev
```

Explained variance Proportion each component

```python
evr=pca13.explained_variance_ratio_
evr
```

Cumulative Explained variance Proportion

```python
cvr=np.cumsum(evr)
cvr
```

```python
sns.lineplot(x=np.arange(1,14),y=evr)
plt.xlabel('Number of components')
plt.ylabel('Explained variance')
plt.title('Elbow Curve')
plt.show()
```

```python
sns.barplot(x=np.arange(1,14),y=cvr)
```

For n_components=6

```python
pca6 = PCA(n_components=6)
pc6 = pca6.fit_transform(X_std)
pc6_df = pd.DataFrame(data = pc6)
```

# Modelling using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
#Splitting the whole data set into Train and Test
X_train, X_test, Y_train, Y_test = train_test_split(X_std,Y, test_size = 0.3, random_state = 5,shuffle=True)
```

Logistic Regression without PCA:

```python
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
yp=model.predict(X_test)
```

```python
model.score(X_test,Y_test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn import metrics
```

```python
print('Accuracy: {:.2f}%'.format(accuracy_score(Y_test,yp) * 100))
print('Classification report:\n\n', classification_report(Y_test, yp))
print('Training set score: {:.2f}%'.format(model.score(X_train, Y_train) * 100))
print('Testing set score: {:.2f}%'.format(model.score(X_test,Y_test) * 100))
```
